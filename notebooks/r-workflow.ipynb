```python
# Remove duplicated values
dataset = unique(dataset)
```

```python
# Remove missing values
dataset = na.omit(dataset)
```

```python
# Check for missing values
colSums(is.na(dataset))
```

## 4. Data Manipulation

```python
# Selecting columns
dataset = dataset %>% select(country, continent, year, lifeExp, pop, gdpPercap)
```

```python
# Renaming columns
dataset = dataset %>% rename(life_exp = lifeExp, gdp_percap = gdpPercap)
head(dataset)
```

```python
# Creating new columns
dataset = dataset %>% mutate(gdp = pop * gdp_percap)
head(dataset)
```

```python
# Filtering rows
dataset = dataset %>% filter(year == 2007)
```

```python
# Sorting rows
dataset = dataset %>% arrange(desc(gdp_percap))
```

## 5. Data Visualization

```python
# Scatterplot
ggplot(dataset, aes(x = gdp_percap, y = life_exp)) + geom_point()
```

```python
# Scatterplot with color
ggplot(dataset, aes(x = gdp_percap, y = life_exp, color = continent)) + geom_point()
```

```python
# Scatterplot with color and size
ggplot(dataset, aes(x = gdp_percap, y = life_exp, color = continent, size = pop)) + geom_point()
```

```python
# Multiple scatterplots
ggplot(dataset, aes(x = gdp_percap, y = life_exp)) + geom_point() + facet_wrap(~continent)
```

```python
# Boxplot
ggplot(dataset, aes(x = continent, y = life_exp)) + geom_boxplot()
```

```python
# Histogram
ggplot(dataset, aes(x = life_exp)) + geom_histogram()
```

```python
# Histogram with color
ggplot(dataset, aes(x = life_exp, fill = continent)) + geom_histogram()
```

```python
# Histogram with color and facet
ggplot(dataset, aes(x = life_exp, fill = continent)) + geom_histogram() + facet_wrap(~continent)
```

```python
# Density plot
ggplot(dataset, aes(x = life_exp, fill = continent)) + geom_density(alpha = 0.5)
```

```python
# Density plot with facet
ggplot(dataset, aes(x = life_exp, fill = continent)) + geom_density(alpha = 0.5) + facet_wrap(~continent)
```

```python
# Animation gapminder
plot <- ggplot(
  dataset,
  aes(x = gdp_percap, y = life_exp, size = pop, color = continent, frame = year)
) +
  geom_point(show.legend = FALSE, alpha = 0.7) +
  scale_color_viridis_d() +
    scale_size_area(max_size = 12) +
    scale_x_log10() +
    labs(title = 'Year: {frame_time}', x = 'GDP per capita', y = 'Life expectancy') +
    theme(legend.position = 'bottom')


plot + transition_time(year) +
  labs(title = "Year: {frame_time}")
```
